```python
# Define catalog and schema
catalog = "dblearning"
schema = "us_accidents"
volume = "raw"

# Define paths
base_path = f"/Volumes/{catalog}/{schema}/{volume}"
raw_data_path = f"{base_path}/"

# Table names
bronze_table_name = f"{catalog}.{schema}.bronze_accidents"
silver_table_name = f"{catalog}.{schema}.silver_accidents"
gold_table_name = f"{catalog}.{schema}.gold_accidents_features"
checkpoint_path = f"{base_path}/_checkpoints/accidents_ingestion"

# Create catalog, schema and volume if it doesn't exist
spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog}.{schema}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.{schema}.{volume}")

# Select the catalog and schema
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE SCHEMA {schema}")

print(f"Setup complete for {catalog}.{schema}")
```
