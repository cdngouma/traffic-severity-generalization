```python
from pyspark.sql.functions import col

# 1. Isolate the "Montreal Proxy" (Boston) - DO NOT SAMPLE THIS
df_boston_test = spark.table("dblearning.us_accidents.silver_accidents").filter(col("City") == "Boston")

# 2. Get the Training Cities (Everything else)
df_train_raw = spark.table("dblearning.us_accidents.silver_accidents").filter(col("City") != "Boston")

# 3. Down-sample only the training data
# We keep all Severity 1, 3, 4 and take 10% of Severity 2
fractions = {1: 1.0, 2: 0.1, 3: 1.0, 4: 1.0}
df_train_balanced = df_train_raw.sampleBy("Severity", fractions, seed=42)

print(f"Boston Test Count: {df_boston_test.count()}")
print(f"Balanced Train Count: {df_train_balanced.count()}")
```
